```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse, parse_qs

# 검색할 키워드와 이미지를 저장할 디렉토리 지정
search_query = "북한장교"  # 원하는 검색어로 변경
output_directory = "downloaded_images"
os.makedirs(output_directory, exist_ok=True)

# 구글 이미지 검색 결과 페이지 URL
url = f"https://www.google.com/search?q={search_query}&tbm=isch"

# 사용자 에이전트 설정
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.61 Safari/537.36"
}

# 구글에 검색 요청을 보내서 HTML 페이지 가져오기
response = requests.get(url, headers=headers)
html = response.text

# BeautifulSoup을 사용하여 HTML 파싱
soup = BeautifulSoup(html, "html.parser")

# 이미지 URL 추출 및 다운로드
image_count = 0
for img in soup.find_all("img", {"data-src": True}):
    img_url = img.get("data-src")
    if img_url and not img_url.startswith("data:image"):
        # 이미지 다운로드
        try:
            response = requests.get(img_url, headers=headers, stream=True)
            if response.status_code == 200:
                # 파일 확장자 추출
                img_extension = os.path.splitext(urlparse(img_url).path)[-1]
                # 이미지 저장
                with open(os.path.join(output_directory, f"image{image_count}{img_extension}"), "wb") as out_file:
                    out_file.write(response.content)
                image_count += 1
        except Exception as e:
            print(f"다운로드 중 오류 발생: {str(e)}")

print(f"{image_count}개의 이미지 다운로드 완료")
```

```python
import os
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from PIL import Image

# 검색할 키워드와 이미지를 저장할 디렉토리 지정
search_query = "북한군복"  # 원하는 검색어로 변경
output_directory = "downloaded_images12"
os.makedirs(output_directory, exist_ok=True)

# Chrome 웹 드라이버 경로 설정 (본인의 드라이버 경로로 수정)
chrome_driver_path = "/path/to/chromedriver"

# Chrome 옵션 설정
chrome_options = Options()
chrome_options.add_argument("--headless")  # 화면 표시 없이 실행 (필요한 경우 제거 가능)

# Chrome 옵션에 executable_path 설정
chrome_options.binary_location = chrome_driver_path

# 웹 드라이버 초기화 및 Chrome 옵션 적용
driver = webdriver.Chrome(options=chrome_options)

# Google 이미지 검색 페이지 열기
driver.get("https://www.google.com/imghp")

# 검색어 입력
search_box = driver.find_element(By.NAME, "q")
search_box.send_keys(search_query)
search_box.send_keys(Keys.RETURN)

# 이미지 스크롤 다운 (원하는 이미지 개수만큼 스크롤 다운)
desired_image_count = 10  # 원하는 이미지 개수로 수정
scroll_count = 0

while scroll_count < desired_image_count:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  # 스크롤 간격 (초)
    scroll_count += 1

# 큰 이미지 URL 추출 및 다운로드 (원하는 크기 이상의 이미지)
min_width = 800  # 원하는 최소 가로 크기로 수정
min_height = 600  # 원하는 최소 세로 크기로 수정
image_count = 0

for img in driver.find_elements(By.CSS_SELECTOR, ".rg_i"):
    img_url = img.get_attribute("src")
    if img_url and not img_url.startswith("data:image"):
        try:
            response = requests.get(img_url, stream=True)
            if response.status_code == 200:
                img_extension = os.path.splitext(urlparse(img_url).path)[-1]
                # 이미지 다운로드 후 크기 확인
                with open(os.path.join(output_directory, f"temp{img_extension}"), "wb") as temp_file:
                    temp_file.write(response.content)
                temp_image = Image.open(os.path.join(output_directory, f"temp{img_extension}"))
                width, height = temp_image.size
                temp_image.close()
                # 최소 크기 이상인 이미지만 저장
                if width >= min_width and height >= min_height:
                    with open(os.path.join(output_directory, f"image{image_count}{img_extension}"), "wb") as out_file:
                        out_file.write(response.content)
                    image_count += 1
                os.remove(os.path.join(output_directory, f"temp{img_extension}"))
        except Exception as e:
            print(f"다운로드 중 오류 발생: {str(e)}")

# 웹 드라이버 종료
driver.quit()

print(f"{image_count}개의 큰 이미지 다운로드 완료")
```
